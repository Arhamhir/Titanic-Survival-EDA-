```python
df = df.drop(columns=['embark_town','deck','embarked'])
```

```python
df['age'].fillna(df['age'].mean(),inplace=True)
df.isna().sum()
```

```python
df.sample(10)
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.select_dtypes(include='number').corr(),annot=True,fmt='.2f',cmap='coolwarm')
plt.show()
```

```python
df_encode = pd.get_dummies(df)
df_encode.astype(int)
df_encode.shape
```

```python
df_encode.columns
```

**using iterative model training to find out the important feaures that are vital for training the model and discarding the rest**

```python
column = ['survived','alive_no','alive_yes','parch','sibsp','pclass','who_child', 'who_man', 'who_woman','sex_female', 'sex_male', 'age']
X = df_encode.drop(columns=column)
y = df_encode['survived']
```

```python
X.sample(10)
```

**columns that the model is being trained on without loosing accuracy** 

```python
for i in df_encode.columns:
    if i not in column:
        print(i)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scale = scaler.fit_transform(X)
X_scale
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_scale,y,random_state=30,test_size=0.20)
```

```python
from lazypredict.Supervised import LazyClassifier
clf = LazyClassifier(verbose=0, ignore_warnings=True)
models, predictions = clf.fit(X_train,X_test,y_train,y_test)
models
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(X_train,y_train)
model.score(X_test,y_test)
model.predict(X_test[0:20])
```

```python
import joblib 
joblib.dump(model,'mymodel.pkl')
joblib.dump(scaler,'myscaler.pkl')
```
